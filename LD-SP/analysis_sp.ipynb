```python
import scipy.signal
import numpy as np
len(test)
```

```python
test=test[['data.A1','timestamp']]
events=scipy.signal.find_peaks(test['data.A1'], height=(2.75), width=1)
# test['data.A1']
```

```python
plt.scatter(events[0], events[1]['peak_heights'], color='r', marker='o')
plt.stem(events[0], events[1]['peak_heights'])
plt.xlim(xmin = 0)
```

```python
test.insert(2,'state',0)
sampleno=0
if len(events[0])>0:
    for x in events[0]:
        if test.loc[x, 'data.A1'] < 5:
            test.at[x, 'state'] = 1
        elif test.loc[x, 'data.A1'] > 9:
            test.at[x, 'state'] = 2
```
