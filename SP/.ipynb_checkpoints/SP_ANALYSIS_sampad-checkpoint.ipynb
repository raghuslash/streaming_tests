```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
#from pylab import rcParams
#rcParams['figure.figsize'] = 15,5
```

```python
import scipy.signal
import numpy as np

%matplotlib notebook
```

```python
SP = pd.read_csv('sp_jan.csv', usecols =['timestamp', 'data.A1'], parse_dates=True)
#only one phase for sp

# Convert to datetime
SP['timestamp']  = pd.to_datetime(SP.timestamp)


window=10
SPstd = SP["data.A1"].rolling(window,center=True,win_type=None).std()
SPmean = SP["data.A1"].rolling(window,center=True,win_type='triang').mean()
SPmax = SP["data.A1"].rolling(window,center=True,win_type=None).max()
```

```python
events=scipy.signal.find_peaks(SP['data.A1'], height=(2.4), width=1)
```

```python
vizFlag=True
```

```python
# stem plots for all peaks
if vizFlag: plt.stem(events[0],events[1]['peak_heights'],'r')
```

```python
# Dist. of all peaks
if vizFlag: sns.distplot(events[1]["peak_heights"])
```

```python
peaks = pd.DataFrame({"sampleNo":events[0],"height":events[1]['peak_heights']})
```

```python
lThresh=5
uThresh=9
realpeaks=peaks[peaks.height.between(lThresh,uThresh).apply(lambda x: not x)]
```

```python
if vizFlag:plt.stem(realpeaks.sampleNo,realpeaks.height,'r')
```

```python
if vizFlag:sns.distplot(realpeaks.height)
```

```python
# Filter SP based on realpeaks indices
SPrealpeaks=SP.ix[realpeaks.sampleNo]
SPrealpeaks["height"] = SPrealpeaks["data.A1"]
```

```python
if vizFlag:
    plt.stem(SPrealpeaks.timestamp,SPrealpeaks.height,'r')
    plt.plot(SP.timestamp,SP["data.A1"])
```

```python
# Filter boards
boards = SPrealpeaks[SPrealpeaks.height < lThresh]

# Filter Cleaning
cleaning = SPrealpeaks[SPrealpeaks.height > uThresh]
```

```python
plt.stem(boards.timestamp,boards.height,'r')
plt.stem(cleaning.timestamp,cleaning.height)
plt.plot(SP.timestamp,SP['data.A1'],'gray')
```
