```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"]=[15,5]
```

```python
raw=pd.read_csv('loader_head.csv')
raw.index = raw["timestamp"]
```

```python
raw["data.acc"] = ( raw["data.ax"]**2 + raw["data.az"]**2 ) ** 0.5
```

```python
raw[["data.gx","data.gy","data.gz"]].plot()
```

```python
raw["data.gyr"]= (raw["data.gy"]**2 +  raw["data.gx"]**2) ** 0.5
```

```python
raw[["data.acc","data.gyr"]].plot()
```

```python
raw.plot.scatter("data.acc","data.gyr",c="black",marker=".")
```

```python
gyroAcc = raw[["data.acc","data.gyr"]]
```

```python
#gyroAcc = gyroAcc.rolling(750).std()
gyroAccNormal = (gyroAcc-gyroAcc.mean())/gyroAcc.std()
#gyroAccNormal=(gyroAcc-gyroAcc.min())/(gyroAcc.max()-gyroAcc.min())
```

```python
gyroAccNormal.plot()
```

```python
(gyroAccNormal["data.acc"] * gyroAccNormal["data.gyr"]).rolling(750).mean().plot()
```

```python
gyroAccNormal.plot.scatter("data.acc","data.gyr",c="r",marker=".")
```
