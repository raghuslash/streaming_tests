```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"]=[15,5]
```

```python
raw=pd.read_csv('loader.csv')
```

```python
acc = ( raw["data.ax"]**2 + raw["data.az"]**2 ) ** 0.5
```

```python
window = 120 * 60 * 75
```

```python
acc.head(window).plot()
```

```python
test=acc.head(window).rolling(75).std()
test.to_csv('rollingstd.csv')
```

```python
acc.head(window).rolling(75).std().head(10000).plot()
```

```python
import scipy.signal
import numpy as np
```

```python
test.plot()
```

```python
refill=scipy.signal.find_peaks(test,height=(0.0075, 0.015),width=25)
```

```python
board=scipy.signal.find_peaks(test,height=(0.02),width=75)
```

```python
refill
```

```python
test.plot()
plt.margins(0)
plt.ylim(ymax = 0.04, ymin = 0)
plt.xlim(xmax = 500000, xmin = 0)
```

```python
plt.stem(board[0], board[1]['peak_heights'])
plt.title('Board Detection')
plt.ylim(ymax = 0.04, ymin = 0)
plt.margins(0)
plt.xlim(xmax = 500000, xmin = 0)
```
